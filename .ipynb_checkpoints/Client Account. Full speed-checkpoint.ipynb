```python
def del_words_regions(arg):
    arg = re.sub(r"[\/\/\\\.,(0)\-']", ' ', str(arg))
    words = re.split(' ', arg)
    new_words = []
    for word in words:
        if word not in stop_words:
            new_words.append(word)
    return ' '.join(new_words)
```

# PCA

## Каменная осыпь

Посик области значений оптимального числа компанент для `PCA`

```python
from sklearn.decomposition import PCA

pca = PCA()
pca.fit(df)
df_pca = pca.transform(df)

# Построение каменной осыпи
sns.set(rc={'figure.figsize':(8,8)})
plt.plot(pca.explained_variance_)
plt.title(label='Каменная осыпь')
plt.xlim(0, 70);
```

Установили, что число компанент лежит в области от 10 до 30

### Определение числа компаненет и число кластеров

```python
from sklearn.cluster import KMeans, AgglomerativeClustering
# для оценки качества кластриезации будет использоваться метрика silhouette_score
# она позволяет оценить качество кластеризации по расстоянию объектов кластера
# что похволяет оценить неразмеченные данные
from sklearn.metrics import silhouette_score

metric_result_list_km = []
metric_result_list_agg = []
for i_companion in range(10, 60, 10):
    pca = PCA(n_components=i_companion)
    pca.fit(df)
    df_pca = pca.transform(df)
    
    silhouette_score_km = []
    silhouette_score_agg = []
    for i in range(2, 70):
        silhouette_score_km.append((silhouette_score(X=df_pca,
                                                      labels=KMeans(n_clusters=i).fit_predict(df_pca)), i))  
        silhouette_score_agg.append((silhouette_score(X=df_pca,
                                                      labels=AgglomerativeClustering(n_clusters=i).fit_predict(df_pca)), i))
    metric_result_list_km.append((i_companion, max(silhouette_score_km)))
    metric_result_list_agg.append((i_companion, max(silhouette_score_agg)))
```

```python
sns.set(rc={'figure.figsize':(8,8)})

#KMeans
plot_km_y, plot_km_x = [i[1][0] for i in metric_result_list_km], [i[0] for i in metric_result_list_km]
plt.plot(plot_km_x, plot_km_y, label='KMeans')
plt.xlabel("n_components for PCA");
plt.ylabel("Silhouette");
plt.title('silhouette_score в зависимости от числа компанент')

#AgglomerativeClustering
plot_agg_y, plot_agg_x = [i[1][0] for i in metric_result_list_agg], [i[0] for i in metric_result_list_agg]
plt.plot(plot_agg_x, plot_agg_y, label='AgglomerativeClustering')
plt.xlabel("n_components for PCA");
plt.ylabel("Silhouette");
plt.title('silhouette_score в зависимости от числа компанент')
plt.legend(['KMeans', 'AgglomerativeClustering']);
```

```python
# оптималное число компанент в PCA для KMeans
def search_max(arg):
    maximum = max([i[1] for i in arg])
    return [i[0] for i in arg if i[1] == maximum][0], maximum[1], maximum[0]

        
optimal_number_components_km, optimal_number_clusters_km, optimal_silhouette_score = search_max(metric_result_list_km)
print('Оптимальное число компанент для KMeans {f1} соответсвующее оптимальному'\
      'числу кластеров {f2},\nс показателем silhouette_score = {f3:.5f}'.format(f1=optimal_number_components_km,
                                                                          f2=optimal_number_clusters_km,
                                                                          f3=optimal_silhouette_score))
```

```python
# оптималное число компанент в PCA для AgglomerativeClustering
optimal_number_components_agg, optimal_number_clusters_agg, optimal_silhouette_score = search_max(metric_result_list_agg)
print('Оптимальное число компанент для AgglomerativeClustering {f1} соответсвующее оптимальному'\
      'числу кластеров {f2},\nс показателем silhouette_score = {f3:.5f}'.format(f1=optimal_number_components_km,
                                                                          f2=optimal_number_clusters_km,
                                                                          f3=optimal_silhouette_score))
```

Стоит обратить внимание, что число компанент имеющая максмальное значение метрики равно 10, а число класетров определено как максимальное число из заданного предела. Из чего можно сделать вывод, что можно попробовать проверить какое число класетров будет оптимально именно для 10 компанент, увеличив число пределы поиска с 70 до 150  

```python
# KMeans

# pca
pca = PCA(n_components=optimal_number_components_km)
pca.fit(df)
df_pca = pca.transform(df)

# predicting
km = KMeans(n_clusters=optimal_number_clusters_km)
predict_km = km.fit_predict(df_pca)

# tsne
from sklearn.manifold import TSNE
tsne = TSNE()
df_tsne = tsne.fit_transform(df_pca)
```

```python
sns.scatterplot(x=df_tsne[:, 0], y=df_tsne[:, 1],
                hue=predict_km)
plt.title(label='{name_alg} number_cluster = {n}'.format(name_alg='KMeans', n=km.n_clusters));
```

```python
# AgglomerativeClustering

# pca
pca = PCA(n_components=optimal_number_components_agg)
pca.fit(df)
df_pca = pca.transform(df)

# predicting
agg = AgglomerativeClustering(n_clusters=optimal_number_clusters_agg)
predict_agg = agg.fit_predict(df_pca)

# tsne
from sklearn.manifold import TSNE
tsne = TSNE()
df_tsne = tsne.fit_transform(df_pca)

# визуализация
sns.scatterplot(x=df_tsne[:, 0], y=df_tsne[:, 1],
                hue=predict_agg)
plt.title(label='{name_alg} number_cluster = {n}'.format(name_alg='AgglomerativeClustering', n=agg.n_clusters));
```

### Поиск оптимального числа класетров

`optimal_number_components_km`, `optimal_number_components_agg` это оптимальное число компанент для PCA
<br>`df_pca` это df к которому было применено pca с оптимальным число компанент
<br> задача, которая стоит сейчас это определить число кластеров

```python
silhouette_score_km = []
silhouette_score_agg = []
for i in range(70, 200):
    silhouette_score_km.append((silhouette_score(X=df_pca,
                                                 labels=KMeans(n_clusters=i).fit_predict(df_pca)), i)) 
    silhouette_score_agg.append((silhouette_score(X=df_pca,
                                                  labels=AgglomerativeClustering(n_clusters=i).fit_predict(df_pca)), i))
```

```python
# KMeans
plot_km_y, plot_km_x = [i[0] for i in silhouette_score_km], [i[1] for i in silhouette_score_km]
plt.plot(plot_km_x, plot_km_y, label='KMeans')


# AgglomerativeClustering
plot_agg_y, plot_agg_x = [i[0] for i in silhouette_score_agg], [i[1] for i in silhouette_score_agg]
plt.plot(plot_agg_x, plot_agg_y, label='AgglomerativeClustering')

# максимальное значение silhouette_score 
optimal_number_clusters_km = max(silhouette_score_km)
optimal_number_clusters_agg = max(silhouette_score_agg)

# метка максимума KMeans
plt.scatter(x=optimal_number_clusters_km[1], y=optimal_number_clusters_km[0], marker='x', c='r')
plt.annotate(s='{label1} - KMeans.\nsil. = {label2}'.format(label1=optimal_number_clusters_km[1],
                                                                     label2=optimal_number_clusters_km[0]),
             xy=(optimal_number_clusters_km[1] + 0.05, optimal_number_clusters_km[0] + 0.0005))

# метка максимума AgglomerativeClustering
plt.scatter(x=optimal_number_clusters_agg[1], y=optimal_number_clusters_agg[0], marker='x', c='r')
plt.annotate(s='{label1} - AgglomerativeCl.\nsil. = {label2}'.format(label1=optimal_number_clusters_agg[1],
                                                                     label2=optimal_number_clusters_agg[0]),
             xy=(optimal_number_clusters_agg[1] + 0.05, optimal_number_clusters_agg[0] + 0.0005))

plt.xlabel("n_clusters");
plt.ylabel("Silhouette");
plt.title('silhouette_score в зависимости от числа компанент')
plt.legend(['KMeans', 'AgglomerativeClustering']);
```

```python
from sklearn.cluster import KMeans, AgglomerativeClustering
# для оценки качества кластриезации будет использоваться метрика silhouette_score
# она позволяет оценить качество кластеризации по расстоянию объектов кластера
# что похволяет оценить неразмеченные данные
from sklearn.metrics import silhouette_score

metric_result_list_km = []
metric_result_list_agg = []
for i_companion in range(10, 60, 10):
    pca = PCA(n_components=i_companion)
    pca.fit(df)
    df_pca = pca.transform(df)
    
    silhouette_score_km = []
    silhouette_score_agg = []
    for i in range(2, 70):
        silhouette_score_km.append((silhouette_score(X=df_pca,
                                                      labels=KMeans(n_clusters=i).fit_predict(df_pca)), i))  
        silhouette_score_agg.append((silhouette_score(X=df_pca,
                                                      labels=AgglomerativeClustering(n_clusters=i).fit_predict(df_pca)), i))
    metric_result_list_km.append((i_companion, max(silhouette_score_km)))
    metric_result_list_agg.append((i_companion, max(silhouette_score_agg)))
```

```python
sns.set(rc={'figure.figsize':(8,8)})

#KMeans
plot_km_y, plot_km_x = [i[1][0] for i in metric_result_list_km], [i[0] for i in metric_result_list_km]
plt.plot(plot_km_x, plot_km_y, label='KMeans')
plt.xlabel("n_components for PCA");
plt.ylabel("Silhouette");
plt.title('silhouette_score в зависимости от числа компанент')

#AgglomerativeClustering
plot_agg_y, plot_agg_x = [i[1][0] for i in metric_result_list_agg], [i[0] for i in metric_result_list_agg]
plt.plot(plot_agg_x, plot_agg_y, label='AgglomerativeClustering')
plt.xlabel("n_components for PCA");
plt.ylabel("Silhouette");
plt.title('silhouette_score в зависимости от числа компанент')
plt.legend(['KMeans', 'AgglomerativeClustering']);
```

```python
# оптималное число компанент в PCA для KMeans
def search_max(arg):
    maximum = max([i[1] for i in arg])
    return [i[0] for i in arg if i[1] == maximum][0], maximum[1], maximum[0]

        
optimal_number_components_km, optimal_number_clusters_km, optimal_silhouette_score = search_max(metric_result_list_km)
print('Оптимальное число компанент для KMeans {f1} соответсвующее оптимальному'\
      'числу кластеров {f2},\nс показателем silhouette_score = {f3:.5f}'.format(f1=optimal_number_components_km,
                                                                          f2=optimal_number_clusters_km,
                                                                          f3=optimal_silhouette_score))
```

```python
# оптималное число компанент в PCA для AgglomerativeClustering
optimal_number_components_agg, optimal_number_clusters_agg, optimal_silhouette_score = search_max(metric_result_list_agg)
print('Оптимальное число компанент для AgglomerativeClustering {f1} соответсвующее оптимальному'\
      'числу кластеров {f2},\nс показателем silhouette_score = {f3:.5f}'.format(f1=optimal_number_components_km,
                                                                          f2=optimal_number_clusters_km,
                                                                          f3=optimal_silhouette_score))
```

Стоит обратить внимание, что число компанент имеющая максмальное значение метрики равно 10, а число класетров определено как максимальное число из заданного предела. Из чего можно сделать вывод, что можно попробовать проверить какое число класетров будет оптимально именно для 10 компанент, увеличив число пределы поиска с 70 до 150  

```python
# KMeans

# pca
pca = PCA(n_components=optimal_number_components_km)
pca.fit(df)
df_pca = pca.transform(df)

# predicting
km = KMeans(n_clusters=optimal_number_clusters_km)
predict_km = km.fit_predict(df_pca)

# tsne
from sklearn.manifold import TSNE
tsne = TSNE()
df_tsne = tsne.fit_transform(df_pca)
```

```python
sns.scatterplot(x=df_tsne[:, 0], y=df_tsne[:, 1],
                hue=predict_km)
plt.title(label='{name_alg} number_cluster = {n}'.format(name_alg='KMeans', n=km.n_clusters));
```

```python
# AgglomerativeClustering

# pca
pca = PCA(n_components=optimal_number_components_agg)
pca.fit(df)
df_pca = pca.transform(df)

# predicting
agg = AgglomerativeClustering(n_clusters=optimal_number_clusters_agg)
predict_agg = agg.fit_predict(df_pca)

# tsne
from sklearn.manifold import TSNE
tsne = TSNE()
df_tsne = tsne.fit_transform(df_pca)

# визуализация
sns.scatterplot(x=df_tsne[:, 0], y=df_tsne[:, 1],
                hue=predict_agg)
plt.title(label='{name_alg} number_cluster = {n}'.format(name_alg='AgglomerativeClustering', n=agg.n_clusters));
```

<br> задача, которая стоит сейчас это определить число кластеров

```python
silhouette_score_km = []
silhouette_score_agg = []
for i in range(70, 350):
    silhouette_score_km.append((silhouette_score(X=df_pca,
                                                 labels=KMeans(n_clusters=i).fit_predict(df_pca)), i)) 
    silhouette_score_agg.append((silhouette_score(X=df_pca,
                                                  labels=AgglomerativeClustering(n_clusters=i).fit_predict(df_pca)), i))
```

```python
# KMeans
plot_km_y, plot_km_x = [i[0] for i in silhouette_score_km], [i[1] for i in silhouette_score_km]
plt.plot(plot_km_x, plot_km_y, label='KMeans')


# AgglomerativeClustering
plot_agg_y, plot_agg_x = [i[0] for i in silhouette_score_agg], [i[1] for i in silhouette_score_agg]
plt.plot(plot_agg_x, plot_agg_y, label='AgglomerativeClustering')

# максимальное значение silhouette_score 
optimal_number_clusters_km = max(silhouette_score_km)
optimal_number_clusters_agg = max(silhouette_score_agg)

# метка максимума KMeans
plt.scatter(x=optimal_number_clusters_km[1], y=optimal_number_clusters_km[0], marker='x', c='r')
plt.annotate(s='{label1} - KMeans.\nsil. = {label2}'.format(label1=optimal_number_clusters_km[1],
                                                                     label2=optimal_number_clusters_km[0]),
             xy=(optimal_number_clusters_km[1] + 0.05, optimal_number_clusters_km[0] + 0.0005))

# метка максимума AgglomerativeClustering
plt.scatter(x=optimal_number_clusters_agg[1], y=optimal_number_clusters_agg[0], marker='x', c='r')
plt.annotate(s='{label1} - AgglomerativeCl.\nsil. = {label2}'.format(label1=optimal_number_clusters_agg[1],
                                                                     label2=optimal_number_clusters_agg[0]),
             xy=(optimal_number_clusters_agg[1] + 0.05, optimal_number_clusters_agg[0] + 0.0005))

plt.xlabel("n_clusters");
plt.ylabel("Silhouette");
plt.title('silhouette_score в зависимости от числа компанент')
plt.legend(['KMeans', 'AgglomerativeClustering']);
```
