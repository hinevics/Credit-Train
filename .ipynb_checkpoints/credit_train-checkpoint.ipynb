```python
list_uniqie_words_from_regions = {re.sub(pattern=r"[\/\/\\\.,(0)\-']",
                                         repl='',
                                         string=j) for www in map(lambda x: x.split(' ') if type(x) is not float else [], regions.values) for j in www}
```

```python
# Город Санкт-Петербург - регион 98
print(list_uniqie_words_from_regions)
stop_words = ['АО', 'КРАЙ', 'РЕСПУБЛИКА', 'Г', 'АОБЛ', 'ОКРУГ', 'ЭЛ', 'ОБЛ', 'РЕСП']
```

```python
def del_words_regions(arg):
    arg = re.sub(r"[\/\/\\\.,(0)\-']", ' ', str(arg))
    words = re.split(' ', arg)
    new_words = []
    for word in words:
        if word not in stop_words:
            new_words.append(word)
    return ' '.join(new_words)
```

```python
regions = regions.map(del_words_regions)
```

### Работа с категориальными данными

```python
df
```

```python
 # имена всех категориальные переменные в одном списке 
name_colums_with_type_object = [i for i in df.columns if df[i].dtype.name == 'object']
```

```python
def fun_insert_new_feath(df, feath, name_feath):
    """
    функция создает на каждый катеориальный признак новый бинарный признак данной катеории
    """
    i_loc = df.columns.get_loc(name_feath)
    for i in feath.columns:
        df.insert(loc=i_loc, column=i, value=feath[i])
    df.drop(columns=[name_feath], inplace=True)
```

```python
name_colums_with_type_object
```

```python
# open_account_flg - целевой признак. его удаляю
name_colums_with_type_object = name_colums_with_type_object[:-1]
name_colums_with_type_object
```

```python
for i in name_colums_with_type_object:
    fun_insert_new_feath(df=df, feath=pd.get_dummies(df[i], prefix=i, dtype='int'), name_feath=i)
```
